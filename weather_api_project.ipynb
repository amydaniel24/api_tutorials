We are turning this into a function called get_weather. It takes in one argumanet (city). This is so we can change the argument without redoing everything. Need to also set funtcion for key.

```python
def get_weather(city):
    url = f"http://api.weatherapi.com/v1/current.json?key={key}&q={city}"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        current = data['current']
        local = data['location']
        temp_f = current['temp_f']
        wind_mph = current['wind_mph']
        humidity = current['humidity']

        print(f"Weather for {local['name']}, {local['country']}:")
        print(f"Temperature: {current['temp_f']}°F")
        print(f"Condition: {current['condition']['text']}")
        print(f"Humidity: {humidity}%")
        print(f"Wind Speed: {wind_mph} mph")

    else:
        print("Error fetching data for (city): (response.status_code). Please check the city name and try again.") 
       
```

We added a for loop and had to make sure to nest the if/else statement inside. Otherwise, it runs but give you the same output as it did above.

```python
cities = ["London", "Mumbai", "New York", "Tokyo", "Sydney"]
weather_data = []

for city in cities:
    url = f"http://api.weatherapi.com/v1/current.json?key={key}&q={city}"
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.json()
        weather_data.append({
            "City": data['location']['name'],
            "Temperature_f": data['current']['temp_f'],
            "Condition": data['current']['condition']['text'],
            "Humidity": data['current']['humidity'],
            "Wind Speed": data['current']['wind_mph']
        })
    else: 
        print("Error fetching data for (city): (response.status_code). Please check the city name and try again.")
```

```python
weather_data
```

It worked, so now we need to create a dataframe with our new list.

```python
weather_df = pd.DataFrame(weather_data)
weather_df.head()
```
