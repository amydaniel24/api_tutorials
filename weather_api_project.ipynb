We are turning this into a function called get_weather. It takes in one argumanet (city). This is so we can change the argument without redoing everything. Need to also set funtcion for key.

```python
def get_weather(city):
    url = f"http://api.weatherapi.com/v1/current.json?key={key}&q={city}"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        current = data['current']
        local = data['location']
        temp_f = current['temp_f']
        wind_mph = current['wind_mph']
        humidity = current['humidity']

        print(f"Weather for {local['name']}, {local['country']}:")
        print(f"Temperature: {current['temp_f']}°F")
        print(f"Condition: {current['condition']['text']}")
        print(f"Humidity: {humidity}%")
        print(f"Wind Speed: {wind_mph} mph")

    else:
        print("Error fetching data for (city): (response.status_code). Please check the city name and try again.") 
       
```

Lets build the function now and try it out.

```python
def get_city_data(city, api_key):

    current_url = f"http://api.weatherapi.com/v1/current.json?key={api_key}&q={city}"
    current_response = requests.get(current_url)

    if current_response.status_code != 200:
        print(f"[CURRENT ERROR] {city} -> {current_response.status_code}")
        print(current_response.text)
        return None

    current_data = current_response.json()

 
    forecast_url = f"http://api.weatherapi.com/v1/forecast.json?key={api_key}&q={city}&days=3"
    forecast_response = requests.get(forecast_url)

    if forecast_response.status_code != 200:
        print(f"[FORECAST ERROR] {city} -> {forecast_response.status_code}")
        print(forecast_response.text)
        return None

    forecast_data = forecast_response.json()

    
    today = pd.Timestamp.now().strftime("%Y-%m-%d")
    astronomy_url = f"http://api.weatherapi.com/v1/astronomy.json?key={api_key}&q={city}&dt={today}"
    astronomy_response = requests.get(astronomy_url)

    if astronomy_response.status_code != 200:
        print(f"[ASTRONOMY ERROR] {city} -> {astronomy_response.status_code}")
        print(astronomy_response.text)
        return None

    astronomy_data = astronomy_response.json()

   
    forecast_days = forecast_data["forecast"]["forecastday"]
    forecast_3_days = []

    for day in forecast_days:
        forecast_3_days.append({
            "Date": day["date"],
            "MaxTemp_F": day["day"]["maxtemp_f"],
            "MinTemp_F": day["day"]["mintemp_f"],
            "AvgTemp_F": day["day"]["avgtemp_f"],
            "Condition": day["day"]["condition"]["text"]
        })

   
    summary = {
        "City": current_data["location"]["name"],
        "Region": current_data["location"]["region"],
        "Country": current_data["location"]["country"],
        "LocalTime": current_data["location"]["localtime"],

        "Temp_F": current_data["current"]["temp_f"],
        "FeelsLike_F": current_data["current"]["feelslike_f"],
        "Condition": current_data["current"]["condition"]["text"],
        "Humidity": current_data["current"]["humidity"],
        "Wind_MPH": current_data["current"]["wind_mph"],

        "Sunrise": astronomy_data["astronomy"]["astro"]["sunrise"],
        "Sunset": astronomy_data["astronomy"]["astro"]["sunset"],
        "MoonPhase": astronomy_data["astronomy"]["astro"]["moon_phase"],

        "Forecast_3_Days": forecast_3_days
    }

    return summary
```

Now we need to make that work for multiple cities at a time. we will have to create a list of cities and a for loop.

```python
multiple_city_data = []

for city in cities:
    print(f"Fetching data for {city}...")
    city_summary = get_city_data(city, api_key)

    if city_summary is not None:
        multiple_city_data.append(city_summary)
```
